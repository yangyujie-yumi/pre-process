# Clustering

## 19.0 Introduction

Frequently, we run into situations where we only know the features.

The goal of clustering algorithms is to identify latent groupings of obesrvations, which if done well, allow us to predict the class of observations even without a target vector.

## 19.1 Clustering Using K-Means

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

iris = datasets.load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = KMeans(n_clusters=3, random_state=0, n_jobs=-1)
model = cluster.fit(features_std)

model.labels_
```

```python
features[:,0]
```

```python
# Dimensionality Reduction for visualization

import matplotlib.pyplot as plt

#Randomly select two column data
L1 = features[:,1]
L2 = features[:,3]

plt.scatter(L1, L2, c=model.labels_, s=50,
            marker='*',cmap='Accent')
```

```python
# Dimensionality Reduction using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # output 2 dims

two_features = pca.fit_transform(features)

# visualization using scatter
plt.scatter(two_features[:,0], two_features[:,1], c=model.labels_, 
            s=50, marker='o',cmap='brg')
```

```python
# visualization using plots

x1, y1 = [], []   
x2, y2 = [], [] 
x3, y3 = [], []

i = 0  
while i < len(two_features):  
    if model.labels_[i]==0:  
        x1.append(two_features[i][0])  
        y1.append(two_features[i][1])  
    elif model.labels_[i]==1:  
        x2.append(two_features[i][0])  
        y2.append(two_features[i][1])  
    elif model.labels_[i]==2:  
        x3.append(two_features[i][0])  
        y3.append(two_features[i][1])
    i = i + 1

# three colors  
plot1, = plt.plot(x1, y1, 'or', marker="o", markersize=5)    
plot2, = plt.plot(x2, y2, 'og', marker="o", markersize=5)    
plot3, = plt.plot(x3, y3, 'ob', marker="o", markersize=5)
plt.title("K-Means Text Clustering")
plt.legend((plot1, plot2, plot3), ('A', 'B', 'C'))

plt.show()
```

### Discussion
k-means clustering is one of the most common clustering techniques. In k-means clustering, the algorithm attempts to group observations into k groups, with each group having roughly equal variance. The number of groups, k, is specified by the user as a hyperparameter. Specifically, in k-means:

1. k cluster "center" points are created at random locations.

2. For each observation:
    a. the distance between each observaiton and the k center points is calculated
    b. the observation is assigned to the cluster of the nearest center point
    
3. The center points are moved to the means (i.e., centers) of their respective clusters

4. Steps 2 and 3 are repeated until no observation changes in cluster membership

k-means clustering assumes:
* the clusters are convex shaped (e.g. a circle, a sphere).
* all features are equally scaled
* the groups are balanced

## 19.2 Speeding Up K-Means Clustering

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans

iris = datasets.load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = MiniBatchKMeans(n_clusters=3, random_state=0, batch_size=100)
model = cluster.fit(features_std)

model.labels_
```

## 19.3 Clustering Using Meanshift

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MeanShift

iris = datasets.load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = MeanShift(n_jobs=-1)
model = cluster.fit(features_std)

model.labels_
```

## 19.4 Clustering Using DBSCAN

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

iris = datasets.load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = DBSCAN(n_jobs=-1)
model = cluster.fit(features_std)

model.labels_
```

## 19.5 Clustering using Hierarchical Merging

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

iris = datasets.load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = AgglomerativeClustering(n_clusters=3)
model = cluster.fit(features_std)

model.labels_
```

### Clustering examples

```python
##Loding data
import pandas as pd

df = pd.read_csv("./datasets/clustering/data_1024.csv")

### For the purposes of this example, we store feature data from our
### dataframe `df`, in the `f1` and `f2` arrays. We combine this into
### a feature matrix `X` before entering it into the algorithm.
f1 = df['Distance_Feature'].values
f2 = df['Speeding_Feature'].values

import matplotlib.pyplot as plt

plt.scatter(f1, f2, s=5,alpha=0.9)
plt.xlabel('Distance_Feature')
plt.ylabel('Speeding_Feature')

plt.figure(figsize=(5,2))
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

X=list(zip(f1,f2))
X = list(map (lambda tup:list(tup), X))
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
assignments = kmeans.labels_
kmeans.predict(X)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(f1, f2, c=kmeans.labels_, s=5,cmap='RdYlBu')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='^', s=100, linewidth=2, c=[0, 1],cmap='Wistia')
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
assignments = kmeans.labels_
```

```python
plt.scatter(f1, f2, c=kmeans.labels_, s=5,cmap='Spectral')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='^', s=100, linewidth=2, c=[0, 1, 3],cmap='copper')
```
